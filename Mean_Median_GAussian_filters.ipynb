```
import numpy as np
import random
import cv2
from google.colab.patches import cv2_imshow
```

```
def sp_noise(image,prob):
    output = np.zeros(image.shape,np.uint8)
    thres = 1 - prob
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            for k in range(image.shape[2]):
                rdn = random.random()
                if rdn < prob:
                    output[i][j][k] = 0
                elif rdn > thres:
                    output[i][j][k] = 255
                else:
                    output[i][j][k] = image[i][j][k]
    return output
 
```

```
image = cv2.imread('SS.jpg') # Only for grayscale image
image.shape
noise_img = sp_noise(image,0.05)
#img = cv2.imread('SS.jpg')
median = cv2.medianBlur(noise_img, 5)
mean = cv2.blur(noise_img,(5,5))
#mean = cv2.meanBlur(noise_img, 5)
#compare = np.concatenate((noise_img, median), axis=1) #side by side comparison
 
cv2_imshow(image)
#cv2_imshow(compare)
cv2_imshow(median)
cv2_imshow(mean)
cv2_imshow(noise_img)
cv2.waitKey(0)
cv2.destroyAllWindows
 

 
```
