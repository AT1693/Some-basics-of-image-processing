```
import cv2
import numpy as np
from matplotlib import pyplot as plt
from scipy import ndimage, misc
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```
img0 = cv2.imread('SS.jpg',)
gray = cv2.cvtColor(img0, cv2.COLOR_BGR2GRAY)
img = cv2.GaussianBlur(gray,(3,3),0)
laplacian = cv2.Laplacian(img,cv2.CV_64F)
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)  # x
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)  # y
ascent = misc.ascent()
result = ndimage.prewitt(ascent)#prewitt

```

```
plt.subplot(2,2,1),plt.imshow(img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,2),plt.imshow(laplacian,cmap = 'gray')
plt.title('Prewitt'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,3),plt.imshow(result,cmap = 'gray')
plt.title('Laplacian'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,4),plt.imshow(sobelx,cmap = 'gray')
plt.title('Sobel X'), plt.xticks([]), plt.yticks([])
#plt.subplot(2,2,5),plt.imshow(sobely,cmap = 'gray')
#plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])
plt.show()
```

```
#after applying canny edges filter
laplacian = laplacian.astype(np.uint8)
sobelx = sobelx.astype(np.uint8)
result = result.astype(np.uint8)
edges1 = cv2.Canny(img,100,200)
edges2 = cv2.Canny(laplacian,100,200)
edges3 = cv2.Canny(sobelx,100,200)
edges4 = cv2.Canny(result,100,200)
cv2_imshow(edges1)
cv2_imshow(edges2)
cv2_imshow(edges3)
cv2_imshow(edges4)
```

```
plt.subplot(2,2,1),plt.imshow(edges1,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,2),plt.imshow(edges2,cmap = 'gray')
plt.title('Prewitt'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,3),plt.imshow(edges3,cmap = 'gray')
plt.title('Laplacian'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,4),plt.imshow(edges4,cmap = 'gray')
plt.title('Sobel X'), plt.xticks([]), plt.yticks([])
```
