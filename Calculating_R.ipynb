```
import numpy as np
import statistics
from scipy import ndimage 
import cv2
from google.colab.patches import cv2_imshow
```

```
img = cv2.imread('pebble.jpg')
img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(img)
x,y=img.shape
l=[]
for i in range(x):
  for j in range(y):
    a=img[i][j]
    l.append(a)
print(l) 
s=sum(l)
mean=0
for i in range(0,x*y):
  mean=(i*(l[i]/s))+mean
print(mean)
sigma=0
l2=l
for i in range (len(l)):
    l2[i]=l2[i]/s
for i in range(len(l2)):
    sigma=(mean-i)**2*l2[i]+sigma
norm=sigma/(len(l2)**2)
r=1-1/(1+norm)
print(r)
```
